## 3. 전처리 + Dataset/DataLoader 구성

앞서 EDA에서 확인한 결과를 기준으로, 이 단계에서는 학습 입력 파이프라인을 고정합니다.

- EDA 핵심 수치: median=101, p95=678, max=78,881(긴 꼬리 분포)
- 전처리 전략: 소문자화, 비알파벳 제거, 불용어 제거, 단일 글자 토큰 제거
- 데이터 분할: Stratified Train/Valid/Test
- 길이 관리: 전처리 후 train 길이분포의 p95를 근거로 `max_len` 결정(상한 캡 적용)
- 벡터화 준비: train 기준 vocab 생성(min_freq=2), PAD/UNK 토큰 정의

이렇게 만든 Dataset/DataLoader는 4단계에서 임베딩별(Word2Vec/FastText/GloVe)로 확장하는 바탕으로 사용합니다.


````python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.datasets import fetch_20newsgroups
from nltk.tokenize import word_tokenize
import nltk

# NLTK tokenizer resources
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)

# 1) Load dataset
news_data = fetch_20newsgroups(
    subset='all',
    remove=('headers', 'footers', 'quotes')
)

texts = news_data.data
labels = news_data.target
target_names = news_data.target_names

print("[데이터 기본 정보]")
print(f"- 전체 문서 수: {len(texts):,}")
print(f"- 클래스 수: {len(set(labels))}")
print(f"- 첫 번째 샘플 라벨: {labels[0]} ({target_names[labels[0]]})")
print("\n[샘플 문서 일부]")
print(texts[0][:800])

# 2) Label distribution
label_counts = Counter(labels)
dist_df = pd.DataFrame({
    'class_id': list(label_counts.keys()),
    'count': list(label_counts.values())
}).sort_values("class_id").reset_index(drop=True)
dist_df["class_name"] = dist_df["class_id"].apply(lambda x: target_names[x])

print("\n[클래스 분포 상위 5개]")
print(dist_df.head())

# 3) Document length distribution
token_lengths = [len(word_tokenize(t)) for t in texts]
series_len = pd.Series(token_lengths)
length_stats = {
    'mean': float(series_len.mean()),
    'median': float(series_len.median()),
    'p95': float(series_len.quantile(0.95)),
    'max': int(series_len.max())
}

print("\n[문서 길이 통계(토큰 기준)]")
for k, v in length_stats.items():
    print(f"- {k}: {v:.2f}" if isinstance(v, float) else f"- {k}: {v}")

plt.figure(figsize=(8, 4))
plt.hist(token_lengths, bins=40, alpha=0.8)
plt.title("20 Newsgroups Document Length Distribution")
plt.xlabel("Number of tokens")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
````

````python
import re
import numpy as np
from collections import Counter
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader
from nltk.corpus import stopwords

nltk.download('stopwords', quiet=True)
STOP_WORDS = set(stopwords.words("english"))

# 1) Stratified split: train/valid/test (72/8/20)
train_texts, test_texts, train_labels, test_labels = train_test_split(
    texts, labels,
    test_size=CFG["test_size"],
    random_state=CFG["random_state"],
    stratify=labels
)

train_texts, valid_texts, train_labels, valid_labels = train_test_split(
    train_texts, train_labels,
    test_size=0.1,
    random_state=CFG["random_state"],
    stratify=train_labels
)

# 2) Text preprocessing
def clean_and_tokenize(text: str):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", " ", text)
    tokens = text.split()
    tokens = [t for t in tokens if t not in STOP_WORDS and len(t) > 1]
    return tokens

train_tokens = [clean_and_tokenize(t) for t in train_texts]
valid_tokens = [clean_and_tokenize(t) for t in valid_texts]
test_tokens  = [clean_and_tokenize(t) for t in test_texts]

# 3) Length-based max_len from preprocessed train tokens
train_lengths = np.array([len(x) for x in train_tokens])
p95_len = int(np.percentile(train_lengths, 95))
MAX_LEN = min(400, max(100, p95_len))

# 4) Build vocabulary on train split only
PAD_TOKEN, UNK_TOKEN = "<PAD>", "<UNK>"
PAD_IDX, UNK_IDX = 0, 1
MIN_FREQ = 2

token_counter = Counter(tok for sent in train_tokens for tok in sent)
vocab_tokens = [tok for tok, cnt in token_counter.items() if cnt >= MIN_FREQ]

word2idx = {PAD_TOKEN: PAD_IDX, UNK_TOKEN: UNK_IDX}
for tok in vocab_tokens:
    word2idx[tok] = len(word2idx)

def encode_tokens(tokens, word2idx, max_len):
    ids = [word2idx.get(t, UNK_IDX) for t in tokens]
    if len(ids) >= max_len:
        return ids[:max_len]
    return ids + [PAD_IDX] * (max_len - len(ids))

class TextIndexDataset(Dataset):
    def __init__(self, token_lists, labels, word2idx, max_len):
        self.token_lists = token_lists
        self.labels = labels
        self.word2idx = word2idx
        self.max_len = max_len

    def __len__(self):
        return len(self.token_lists)

    def __getitem__(self, idx):
        x = encode_tokens(self.token_lists[idx], self.word2idx, self.max_len)
        y = int(self.labels[idx])
        return torch.tensor(x, dtype=torch.long), torch.tensor(y, dtype=torch.long)

# 5) Dataset/DataLoader
BATCH_SIZE = 64
train_dataset = TextIndexDataset(train_tokens, train_labels, word2idx, MAX_LEN)
valid_dataset = TextIndexDataset(valid_tokens, valid_labels, word2idx, MAX_LEN)
test_dataset = TextIndexDataset(test_tokens, test_labels, word2idx, MAX_LEN)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# 6) Sanity check
xb, yb = next(iter(train_loader))
print("[데이터 분할 크기]")
print(f"- Train: {len(train_dataset):,}")
print(f"- Valid: {len(valid_dataset):,}")
print(f"- Test : {len(test_dataset):,}")
print("\n[전처리 후 길이 통계(Train)]")
print(f"- mean: {train_lengths.mean():.2f}")
print(f"- median: {np.median(train_lengths):.2f}")
print(f"- p95: {p95_len}")
print(f"- chosen MAX_LEN: {MAX_LEN}")
print(f"\n[단어집 크기] {len(word2idx):,}")
print(f"[배치 shape] x={tuple(xb.shape)}, y={tuple(yb.shape)}")
````

## 4. 임베딩별 모델 학습(Word2Vec / FastText / GloVe)

3번 단계 출력값을 기준으로 학습 실험을 진행합니다.
- 입력 길이: `MAX_LEN` (3번에서 산정된 값, 현재 실행 예시: 288)
- 단어집: `word2idx` (3번에서 생성된 vocab, 현재 실행 예시: 42,203)
- 데이터 분할: train/valid/test 고정
- 비교 방식: 임베딩만 바꾸고 모델/학습 하이퍼파라미터는 동일하게 유지

각 임베딩별로:
1) 임베딩 행렬 생성
2) 동일한 RNN 분류기 학습(Valid F1-macro 기준 best 저장)
3) Test에서 Accuracy / Precision / Recall / F1 비교


````python
import time
import copy
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from IPython.display import display
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from gensim.models import Word2Vec, FastText
import gensim.downloader as api

print(f"[Step3 linkage] MAX_LEN={MAX_LEN}, vocab_size={len(word2idx):,}")

# ------------------------------
# 1) Embedding resources
# ------------------------------
EMBED_DIM = 100
W2V_EPOCHS = 10

print("\n[Embedding train/load]")
start_embed = time.time()

word2vec_model = Word2Vec(
    sentences=train_tokens,
    vector_size=EMBED_DIM,
    window=5,
    min_count=1,
    sg=1,
    workers=4,
    epochs=W2V_EPOCHS,
    seed=SEED
)

fasttext_model = FastText(
    sentences=train_tokens,
    vector_size=EMBED_DIM,
    window=5,
    min_count=1,
    sg=1,
    workers=4,
    epochs=W2V_EPOCHS,
    seed=SEED
)

try:
    glove_kv = api.load("glove-wiki-gigaword-100")
except Exception as e:
    raise RuntimeError(f"GloVe loading failed: {e}. Check network/runtime.")

print(f"- embedding resources ready ({time.time() - start_embed:.1f}s)")

def build_embedding_matrix(kv_like, word2idx, embed_dim, pad_idx=0):
    matrix = np.random.normal(loc=0.0, scale=0.02, size=(len(word2idx), embed_dim)).astype(np.float32)
    matrix[pad_idx] = 0.0
    covered = 0
    for word, idx in word2idx.items():
        if idx == pad_idx:
            continue
        try:
            vec = kv_like[word]
        except KeyError:
            continue
        if vec.shape[0] != embed_dim:
            continue
        matrix[idx] = vec
        covered += 1
    coverage = covered / max(1, len(word2idx) - 1)
    return matrix, coverage

embedding_matrices = {}
coverages = {}

embedding_matrices["word2vec"], coverages["word2vec"] = build_embedding_matrix(word2vec_model.wv, word2idx, EMBED_DIM, PAD_IDX)
embedding_matrices["fasttext"], coverages["fasttext"] = build_embedding_matrix(fasttext_model.wv, word2idx, EMBED_DIM, PAD_IDX)
embedding_matrices["glove"], coverages["glove"] = build_embedding_matrix(glove_kv, word2idx, EMBED_DIM, PAD_IDX)

print("\n[Embedding coverage]")
for name in ["word2vec", "fasttext", "glove"]:
    print(f"- {name}: {coverages[name]*100:.2f}%")

````

````python
# ------------------------------
# 2) Shared classifier
# ------------------------------
class EmbeddingRNNClassifier(nn.Module):
    def __init__(self, embedding_matrix, hidden_dim=128, num_layers=1, dropout=0.3):
        super().__init__()
        _, embedding_dim = embedding_matrix.shape
        self.embedding = nn.Embedding.from_pretrained(
            torch.tensor(embedding_matrix, dtype=torch.float32),
            freeze=False,
            padding_idx=PAD_IDX
        )
        rnn_type = CFG.get("rnn_model", "lstm").lower()
        self.rnn_type = rnn_type
        rnn_cls = nn.LSTM if rnn_type == "lstm" else nn.GRU
        self.rnn = rnn_cls(
            input_size=embedding_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            bidirectional=True,
            dropout=dropout if num_layers > 1 else 0.0
        )
        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(hidden_dim * 2, CFG["num_classes"])

    def forward(self, x):
        x = self.embedding(x)
        if self.rnn_type == "lstm":
            _, (h, _) = self.rnn(x)
        else:
            _, h = self.rnn(x)
        h_last = torch.cat([h[-2], h[-1]], dim=1)
        out = self.fc(self.dropout(h_last))
        return out

def train_one_epoch(model, loader, criterion, optimizer, device):
    model.train()
    total_loss = 0.0
    for x, y in loader:
        x, y = x.to(device), y.to(device)
        optimizer.zero_grad()
        logits = model(x)
        loss = criterion(logits, y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(loader)

def evaluate(model, loader, criterion, device):
    model.eval()
    total_loss = 0.0
    y_true, y_pred = [], []
    with torch.no_grad():
        for x, y in loader:
            x, y = x.to(device), y.to(device)
            logits = model(x)
            loss = criterion(logits, y)
            total_loss += loss.item()
            pred = torch.argmax(logits, dim=1)
            y_true.extend(y.cpu().numpy().tolist())
            y_pred.extend(pred.cpu().numpy().tolist())

    acc = accuracy_score(y_true, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(
        y_true, y_pred, average="macro", zero_division=0
    )
    return {
        "loss": total_loss / len(loader),
        "accuracy": acc,
        "precision_macro": prec,
        "recall_macro": rec,
        "f1_macro": f1,
    }

````

````python
# ------------------------------
# 3) Train/Eval loop per embedding
# ------------------------------
LR = 1e-3
EPOCHS = 5
PATIENCE = 2

results = []

for emb_name in ["word2vec", "fasttext", "glove"]:
    print(f"\n===== [{emb_name}] training start =====")
    emb_matrix = embedding_matrices[emb_name]

    model = EmbeddingRNNClassifier(embedding_matrix=emb_matrix, hidden_dim=128, num_layers=1, dropout=0.3).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=LR)

    best_f1 = -1.0
    best_state = None
    no_improve = 0

    for epoch in range(1, EPOCHS + 1):
        tr_loss = train_one_epoch(model, train_loader, criterion, optimizer, device)
        va = evaluate(model, valid_loader, criterion, device)
        print(
            f"[{emb_name}] epoch {epoch:02d} "
            f"train_loss={tr_loss:.4f} "
            f"valid_loss={va['loss']:.4f} "
            f"valid_f1={va['f1_macro']:.4f}"
        )

        if va["f1_macro"] > best_f1:
            best_f1 = va["f1_macro"]
            best_state = copy.deepcopy(model.state_dict())
            no_improve = 0
        else:
            no_improve += 1
            if no_improve >= PATIENCE:
                print(f"[{emb_name}] early stopping")
                break

    model.load_state_dict(best_state)
    test_metrics = evaluate(model, test_loader, criterion, device)
    test_metrics["embedding"] = emb_name
    test_metrics["coverage"] = coverages[emb_name]
    results.append(test_metrics)

    print(
        f"[{emb_name}] TEST "
        f"acc={test_metrics['accuracy']:.4f}, "
        f"prec={test_metrics['precision_macro']:.4f}, "
        f"rec={test_metrics['recall_macro']:.4f}, "
        f"f1={test_metrics['f1_macro']:.4f}"
    )

````

## 6. 성능 개선 실험 및 최종 결론

5단계 결과에서 선정된 최상위 임베딩을 기준으로, 간단한 하이퍼파라미터 튜닝을 시도합니다.

1) 기준 모델: 5단계의 best embedding 성능(보통 F1-macro 기준)
2) 튜닝 대상: hidden size, dropout, learning rate, epoch
3) 선정 기준: Valid F1-macro
4) 결과 산출: baseline vs tuned 비교표 + 최종 함수 저장


````python
import copy
import time
from pathlib import Path
import pandas as pd
import torch
import torch.nn as nn
from IPython.display import display

# prerequisites check
required_vars = [
    "result_df", "embedding_matrices", "train_loader", "valid_loader", "test_loader",
    "EmbeddingRNNClassifier", "train_one_epoch", "evaluate", "device"
]
missing = [v for v in required_vars if v not in globals()]
if missing:
    raise RuntimeError(f"Run Sections 3, 4, and 5 first. Missing: {missing}")

# 1) baseline from section 5/4
base_df = result_df.sort_values("f1_macro", ascending=False).reset_index(drop=True).copy()
best_embedding = base_df.loc[0, "embedding"]
baseline_metrics = base_df.loc[0, ["accuracy", "precision_macro", "recall_macro", "f1_macro", "loss"]].to_dict()

print("[Section 6] Tuning around best embedding")
print(f"- Best embedding from Section 5: {best_embedding}")
print(f"- Baseline F1-macro: {baseline_metrics['f1_macro']:.4f}")

# 2) hyperparameter candidates
trial_configs = [
    {"hidden_dim": 128, "dropout": 0.30, "lr": 1e-3, "num_layers": 1, "epochs": 5},
    {"hidden_dim": 192, "dropout": 0.30, "lr": 1e-3, "num_layers": 1, "epochs": 5},
    {"hidden_dim": 128, "dropout": 0.50, "lr": 8e-4, "num_layers": 1, "epochs": 6},
    {"hidden_dim": 192, "dropout": 0.50, "lr": 8e-4, "num_layers": 1, "epochs": 6},
]

PATIENCE = 2
emb_matrix = embedding_matrices[best_embedding]
trial_records = []

for t_idx, cfg_t in enumerate(trial_configs, start=1):
    print(f"\n--- Trial {t_idx}/{len(trial_configs)}: {cfg_t}")
    model = EmbeddingRNNClassifier(
        embedding_matrix=emb_matrix,
        hidden_dim=cfg_t["hidden_dim"],
        num_layers=cfg_t["num_layers"],
        dropout=cfg_t["dropout"],
    ).to(device)

    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=cfg_t["lr"])

    best_state = None
    best_valid = None
    best_valid_f1 = -1.0
    no_improve = 0

    t0 = time.time()
    for epoch in range(1, cfg_t["epochs"] + 1):
        tr_loss = train_one_epoch(model, train_loader, criterion, optimizer, device)
        va = evaluate(model, valid_loader, criterion, device)
        print(
            f"  epoch {epoch:02d} "
            f"train_loss={tr_loss:.4f} "
            f"valid_f1={va['f1_macro']:.4f}"
        )

        if va["f1_macro"] > best_valid_f1:
            best_valid_f1 = float(va["f1_macro"])
            best_valid = va
            best_state = copy.deepcopy(model.state_dict())
            no_improve = 0
        else:
            no_improve += 1
            if no_improve >= PATIENCE:
                print("  early stopping")
                break

    model.load_state_dict(best_state)
    te = evaluate(model, test_loader, criterion, device)
    elapsed = time.time() - t0

    trial_records.append({
        "trial_id": t_idx,
        "embedding": best_embedding,
        "hidden_dim": cfg_t["hidden_dim"],
        "dropout": cfg_t["dropout"],
        "lr": cfg_t["lr"],
        "epochs": cfg_t["epochs"],
        "valid_f1_macro": best_valid["f1_macro"],
        "test_accuracy": te["accuracy"],
        "test_precision_macro": te["precision_macro"],
        "test_recall_macro": te["recall_macro"],
        "test_f1_macro": te["f1_macro"],
        "test_loss": te["loss"],
        "elapsed_sec": elapsed,
        "state_dict": copy.deepcopy(best_state),
    })

# 3) select tuned best by valid f1, then compare with baseline
tuning_df = pd.DataFrame([{k: v for k, v in r.items() if k != "state_dict"} for r in trial_records])
tuning_df = tuning_df.sort_values("valid_f1_macro", ascending=False).reset_index(drop=True)
display(tuning_df)

best_trial = max(trial_records, key=lambda r: r["valid_f1_macro"])
improve = best_trial["test_f1_macro"] - baseline_metrics["f1_macro"]

comparison_df = pd.DataFrame([
    {
        "model": "baseline(best embedding)",
        "accuracy": baseline_metrics["accuracy"],
        "precision_macro": baseline_metrics["precision_macro"],
        "recall_macro": baseline_metrics["recall_macro"],
        "f1_macro": baseline_metrics["f1_macro"],
        "loss": baseline_metrics["loss"],
    },
    {
        "model": "tuned(best trial)",
        "accuracy": best_trial["test_accuracy"],
        "precision_macro": best_trial["test_precision_macro"],
        "recall_macro": best_trial["test_recall_macro"],
        "f1_macro": best_trial["test_f1_macro"],
        "loss": best_trial["test_loss"],
    },
])

print("\n[Baseline vs Tuned]")
display(comparison_df)
print(f"- F1 improvement (tuned - baseline): {improve:+.4f}")

# 4) save final tuned checkpoint
save_dir = Path("outputs")
save_dir.mkdir(parents=True, exist_ok=True)
ckpt_path = save_dir / "mission10_best_tuned.pt"

torch.save({
    "embedding": best_embedding,
    "trial_config": {
        "hidden_dim": best_trial["hidden_dim"],
        "dropout": best_trial["dropout"],
        "lr": best_trial["lr"],
        "num_layers": 1,
        "max_len": MAX_LEN,
    },
    "word2idx": word2idx,
    "model_state_dict": best_trial["state_dict"],
    "baseline_metrics": baseline_metrics,
    "tuned_test_metrics": {
        "accuracy": best_trial["test_accuracy"],
        "precision_macro": best_trial["test_precision_macro"],
        "recall_macro": best_trial["test_recall_macro"],
        "f1_macro": best_trial["test_f1_macro"],
        "loss": best_trial["test_loss"],
    },
}, ckpt_path)

print(f"- Saved tuned checkpoint: {ckpt_path}")
if improve > 0:
    print("- Final conclusion: tuned model selected for submission.")
else:
    print("- Final conclusion: baseline model retained (no meaningful gain from tuning).")
````
